```python
df = pd.read_csv("data.csv", encoding="ISO-8859-1")
```

```python
df.head(5)
```

```python
df.columns=df.columns.str.strip()
print(df.columns)
```

**DATA CLEANING AND MANIPULATION**

```python
print(df.isnull().values.any())
```

```python
df["Description"] = df["Description"].fillna("Unknown") #replace missing decription with Unknown
```

```python
df['CustomerID']=df['CustomerID'].fillna(-1)
```

```python
df[df.isnull().any(axis=1)] # to know which rows have missing values
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

```python
df.head(3)
```

```python
print(df.dtypes)
```

```python
df["Quantity"] = pd.to_numeric(df["Quantity"])
df["UnitPrice"] = pd.to_numeric(df["UnitPrice"])
```

```python
print(df[["Quantity", "UnitPrice"]].isnull().sum())
```

```python
df["TotalSales"]= (df['Quantity']* df["UnitPrice"]).round(2)
```

```python
df.head(3)
```

```python
df.duplicated().sum() #identify any duplicate rows
```

```python
df = df.drop_duplicates() #drop duplicates
```

```python
df.info()
```

```python
df["IsCancelled"] = df["InvoiceNo"].astype(str).str.startswith("C")
print(df["IsCancelled"].value_counts())
```

```python
df["IsValidSale"] = (
    (df["Quantity"] > 0) &
    (df["UnitPrice"] > 0) &
    (df["TotalSales"] > 0) &
    (df["IsCancelled"] == False)
)
```

```python
print(df["IsValidSale"].value_counts())
```

```python
df["CustomerID"] = df["CustomerID"].fillna(-1).astype(int)
```

```python
df
```

```python
df.to_csv("cleaned_dataset.csv",index=False)
```
